```python
# 02_exploratory_data.ipynb

import sys
sys.path.append('../scripts')

from eda import describe_data, show_data

df = show_data('../data/cleaned_data.csv')
```

```python
describe_data(df)
```

Key Features:

1. User ID: Unique identifier for each user.  
2. Device Model: Model of the user's smartphone.  
3. Operating System: The OS of the device (iOS or Android).  
4. App Usage Time: Daily time spent on mobile applications, measured in minutes.  
5. Screen On Time: Average hours per day the screen is active.  
6. Battery Drain: Daily battery consumption in mAh.  
7. Number of Apps Installed: Total apps available on the device.  
8. Data Usage: Daily mobile data consumption in megabytes.  
9. Age: Age of the user.  
10. Gender: Gender of the user (Male or Female).  
11. User Behavior Class: Classification of user behavior based on usage patterns (1 to 5).  

#Visualization of the frequency of the screen time and the frequency of the app usage time with gender comparison

The number of males collected on this data is only 8,3% more than the females. so we can assume that the gender imbalance is not due to the data collection.

On average people spends around 5 hours on the phone per day.

The highest frequency for both genders appears to be up to 100 minutes at phone usage time. Overall, The overlaid line graph suggests a general declining trend in frequency as usage time increases, with some spikes at certain intervals.

```python
from eda import plot_histogram, plot_bar_chart, plot_scatter_plot, plot_boxplot, plot_heatmap

plot_histogram(df, ['App Usage Time (min/day)', 'Screen On Time (hours/day)'], 'Gender')

plot_bar_chart(df,'Gender')
```

#Visualization of the relationship between phone types and gender.

The data shows that there's a slight more popularity of females preferring the iPhone more than males. While Android is more popular between males and overall.

```python
plot_bar_chart(df,'Device Model', 'Gender')
plot_bar_chart(df,'Operating System', 'Gender')
```

Between ages, Iphone are more popular between older people and people in their 30s.

```python
plot_histogram(df,['Age'], 'Gender', bins=4)

plot_histogram(df,['Age'], 'Device Model', bins=4)
```

#Visualization of data usage

The Data Usage representation shows that the maojority of the people doesn't use much of the mobile internet. There's a peak on less than 500 MB/day of data consumed and the tendency is to decline.

```python
plot_histogram(df, ['Data Usage (MB/day)'], 'Gender')
```

#Visualization of user behavior class relationships.

The highest user behavior class more apps are installed in their device and more mobile data are used. could be useful for user segmentation, understanding user engagement patterns, or tailoring app recommendations based on user behavior class. It provides insights into how app installation habits relate to daily device usage across different user types.

```python
plot_scatter_plot(df, 'Number of Apps Installed', 'Data Usage (MB/day)', 'User Behavior Class')
plot_scatter_plot(df, 'Number of Apps Installed', 'App Usage Time (min/day)', 'User Behavior Class')
plot_scatter_plot(df, 'Number of Apps Installed', 'Screen On Time (hours/day)', 'User Behavior Class')
plot_scatter_plot(df, 'Battery Drain (mAh/day)', 'App Usage Time (min/day)', 'User Behavior Class')
```

According to the data android devices tend to be more battery efficient than iOS devices although the difference is not significant.

```python
plot_boxplot(df, 'Device Model', 'Battery Drain (mAh/day)')
```

```python
plot_boxplot(df, 'Operating System', 'Battery Drain (mAh/day)')
```

This correlation matrix reveals that app usage time, screen time, battery drain, number of apps installed, and data usage are closely related behaviors. The User Behavior Class seems to be a classification based on these usage patterns. Demographic factors like age and gender don't appear to have strong relationships with usage behaviors in this dataset. This information could be valuable for app developers, device manufacturers, or researchers studying smartphone usage patterns.

```python
# 02_exploratory_data.ipynb
import pandas as pd
import sys
sys.path.append('../scripts')

data = pd.read_csv('../data/market_sales_cleaned.csv')
data.head(20)
```

```python
from eda import demographics_graph, stats, price_avg_graph, sales_quantity_graph, revenue_graph, sales_over_time_graph, rating_by_gender

demographics_graph(data)

#The difference between the number of males and females is very small.
#The distribution of customer type is very balanced with members being slight bigger than normal. With more females in member customer type than normal customer type
```

```python
stats(data, 'Unit price')
```

```python
stats(data, 'Total')
```

```python
price_avg_graph(data)
```

```python
data.groupby('Product line').agg({'Unit price': 'mean', 'Quantity': 'sum'}).reset_index()
```

```python
sales_quantity_graph(data)
```

```python
revenue_graph(data, 'Gender')

#"Food and beverages" and "Sport and travel" are the top sales products, followed by "Electronic accessories" and "Fashion accessories"
#"Health and beauty" shows the lowest performance of all products.
#Females customers are the main clients across most products lines specially on "Food and beverages".
#"Electronic accessories" is balanced between males and females.
#"Health and beauty" is were the Males are the majority of consumers.
```

```python
data.groupby('Product line').agg({'Total': 'sum'}).sort_values(by='Total', ascending=False).reset_index()
```

```python
revenue_graph(data, 'Customer type')

#In general there's a slight more sales from members than normal customers.
#There's more sales on "Fashion accessories" and "Electronic accessories" from normal customers than members and vice-versa on the rest of the products.
```

```python
data.groupby(['Customer type']).agg({'Total': 'sum'}).sort_values(by='Total', ascending=False).reset_index()
```

```python
revenue_graph(data, 'City')

#"Yangon" is the city with the highest sales, followed by "Naypyitaw" and "Mandalay".
#"Food and beverages" and "Fashion accessories" are the top categories of products sold in the city.
#Naypyitaw shows notable strength in Food and beverages and Fashion accessories.
#Yangon shows strong performance in Home and lifestyle products.
#Mandalay appears strongest in Health and beauty.
```

```python
data.groupby(['City']).agg({'Total': 'sum'}).sort_values(by='Total', ascending=False).reset_index()
```

```python
revenue_graph(data, 'Branch')

#This graph behaviour is the same as the revenue by city, showing which branches are the main stores for each city.
```

```python
sales_over_time_graph(data)

#the data only has 3 months worth of data, so the graph is not very informative despite showing declination in revenue over time.
```

```python
rating_by_gender(data)

#There's no significant difference in the ratings distribution between genders. Which suggests that the rating is not influenced by the gender of the customer.
#Therefore, there's no need for improvement strategies based on gender for customer satisfaction.
```
